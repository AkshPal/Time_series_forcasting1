```python
df = pd.read_excel("nifty50.xlsx",index_col='date',parse_dates = True)
df.dropna()
df.head(5)
```

```python
df.describe()
print(df.shape)
```

```python
df.plot()
```

```python
from statsmodels.tsa.stattools import adfuller

res = adfuller(df.close.dropna(),autolag='AIC')
print(f"ADF Statistic: {res[0]}")
print(f"p-value:{res[1]}")
print(f"lags:{res[2]}")
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf



acf_or = plot_acf(df.close)
pcf_or = plot_pacf(df.close)
```

```python
df_diff = df.close.diff().dropna()
df_diff.plot()
```

```python
acf_diff = plot_acf(df_diff)
pacf_diff = plot_pacf(df_diff)
```

```python
res_diff= adfuller(df_diff.dropna(),autolag='AIC')
print(f"ADF Statistic: {res_diff[0]}")
print(f"p-value:{res_diff[1]}")
print(f"lags:{res_diff[2]}")
```

```python
from pmdarima.arima.utils import ndiffs
```

```python
ndiffs(df.dropna(),test="adf")
```

```python
from statsmodels.tsa.arima_model import ARIMA

from pmdarima import auto_arima

import warnings

warnings.filterwarnings("ignore")

# model = auto_arima(df.dropna(),trace = True,supress_warnings = True)
model = auto_arima(df.dropna(), start_p=1, start_q=1,
                         test='adf',
                         max_p=5, max_q=5, m=12,
                         start_P=0, seasonal=True,
                         d=None, D=1, trace=True,
                         error_action='ignore',  
                         suppress_warnings=True, 
                         stepwise=True)

model.summary()
```

```python
train = df.iloc[:-15]
test = df.iloc[-15:]
print(train.shape,test.shape)
```

```python
from statsmodels.tsa.arima.model import ARIMA
model_=ARIMA(train['close'],order=(2,0,2),seasonal_order=(2,1,0,12))
model_=model_.fit()
model_.summary()
```

```python
st = len(train)
ed = len(train)+len(test)-1
pr = model_.predict(start = st,end = ed,type='levels')
pr.index=df.index[st:ed+1]
print(pr)
```

```python
pr.plot(legend=True)
test.plot(legend=True)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rmse = sqrt(mean_squared_error(pr,test))
print(rmse)
```
